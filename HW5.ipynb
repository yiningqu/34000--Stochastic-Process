```python
import random
import matplotlib.pyplot as plt
```

```python
result_2000 = []
result_4000 = []
for j in range(30):
    R = 1
    G = 1
    for i in range(4050):
        values = [0, 1]
        probabilities = [R/(R+G), G/(R+G)]  # 30% chance of 0, 70% chance of 1

        random_choice = random.choices(values, probabilities)[0]
        if random_choice == 0:
            R += 1
        else:
            G += 1
        M = R / (i + 2)

        if i == 1999:
            #print("For 2000 draws, the fraction is " + str(M))
            result_2000.append(M)
        if i == 3999:
            #print("For 4000 draws, the fraction is " + str(M))
            result_4000.append(M)

plt.plot(result_2000)
plt.plot(result_4000)
plt.legend([2000,4000])
plt.show


```
